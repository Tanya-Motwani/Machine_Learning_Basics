```

import numpy
from sklearn import datasets, metrics
from sklearn.naive_bayes import GaussianNB
import matplotlib.pyplot as plt
```

```
iris=datasets.load_iris()

trainingset= iris.data[range(1,150,2),:]
trainingsettarget=iris.target[range(1,150,2)]

testingset= iris.data[range(0,150,2),:]
testsetarget=iris.target[range(0,150,2)]
```

```
plt.plot(iris.data[:,1])
```

```
clf=GaussianNB()
#MutliBinomialNB()
#BernoulliNB()

clf.fit(trainingset, trainingsettarget)
predictions=clf.predict(testingset)
```

```
print("Accuracy={0}".format(metrics.accuracy_score(predictions,testsetarget,normalize=True)))

print(metrics.classification_report(predictions,testsetarget))

print(metrics.confusion_matrix(predictions,testsetarget))
```
